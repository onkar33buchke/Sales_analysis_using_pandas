### Importing Libraries

```python
import numpy as np
import pandas as pd
import os
```

### Merging the 12 month sales data into a single csv file

```python
df = pd.read_csv(".\Sales_data_monthwise_csv\Sales_January_2019.csv")
files = [file for file in os.listdir(".\Sales_data_monthwise_csv")]

all_months_data = pd.DataFrame()
```

```python
for file in files:
    df = pd.read_csv(".\Sales_data_monthwise_csv\\" +file)
    all_months_data = pd.concat([all_months_data,df])
    
all_months_data.head()
```

### Export combined dataframe to new csv

```python
all_months_data.to_csv('Combined_all_months_data.csv',index=False)
```

#### Read in updated dataframe

```python
all_data = pd.read_csv(".\Combined_data_csv\Combined_all_months_data.csv")

all_data.head()
```

### Cleaning the data

#### Remove rows with all NaN

```python
nan_df = all_data[all_data.isna().any(axis=1)]
nan_df.head()
```

```python
all_data = all_data.dropna(how='all')
all_data.head()
```

#### Find date months (from 'Order Date' column) that starts with 'Or' and delete it

```python
temp_df = all_data[all_data['Order Date'].str[0:2] == 'Or']
temp_df.head()
```

```python
all_data = all_data[all_data['Order Date'].str[0:2] != 'Or']
all_data.head()
```

#### Convert columns to the correct type

```python
all_data['Quantity Ordered'] = pd.to_numeric(all_data['Quantity Ordered'])
all_data['Price Each'] = pd.to_numeric(all_data['Price Each'])
```

### Adding additional columns to dataframe

#### Add month column

```python
all_data['Month'] = all_data['Order Date'].str[0:2]
all_data['Month'] = all_data['Month'].astype('int32')

all_data.head()
```

#### Add sales column

```python
all_data['Sales'] = all_data['Quantity Ordered'] * all_data['Price Each']

all_data.head()
```

#### Add city column

```python
def get_city(address):
    return address.split(',')[1]

def get_state(address):
    return address.split(',')[2].split(' ')[1]
all_data['City'] = all_data['Purchase Address'].apply(lambda x: get_city(x) + " (" + get_state(x) + ")")

#other method
#all_data['City'] = all_data['Purchase Address'].apply(lambda x: f"{get_city(x)} ({get_state(x)})")

all_data.head()
```

### What was the best month for sales? How much was earned that month?

```python
results = all_data.groupby('Month').sum()
```

```python
import matplotlib.pyplot as plt

months = range(1,13)

plt.bar(months,results['Sales'])
plt.xticks(months)
plt.ylabel('Sales in USD ($)')
plt.xlabel('Month Number')
plt.show()
```

### What city had the highest number of sales?

```python
results = all_data.groupby('City').sum()

results
```

```python
import matplotlib.pyplot as plt

from matplotlib.figure import Figure
fig = plt.figure(figsize =(8, 6))

cities = [city for city, df in all_data.groupby('City')]

plt.bar(cities,results['Sales'])
plt.xticks(cities,rotation='vertical',size=8)
plt.ylabel('Sales in USD ($)')
plt.xlabel('cities')
plt.ylim([1000000,9000000])
plt.show()
```

### What time should we display advertisements to maximze likelihood of customer's buying product?

```python
all_data['Order Date'] = pd.to_datetime(all_data['Order Date'])
```

```python
all_data['Hour'] = all_data['Order Date'].dt.hour
all_data['Minute'] = all_data['Order Date'].dt.minute

all_data.head()
```

```python
hours = [hour for hour, df in all_data.groupby('Hour')]

plt.plot(hours,all_data.groupby(['Hour']).count())

plt.xticks(hours)
plt.xlabel('Hour')
plt.ylabel("Number of Orders")
plt.grid()
plt.show()
```

### What products are most often sold together?

```python
df11 = all_data[all_data['Order ID'].duplicated(keep=False)]

df11['Grouped'] = df11.groupby('Order ID')['Product'].transform(lambda x: ",".join(x))

df11 = df11[['Order ID','Grouped']].drop_duplicates()

df11.head(100)
```

```python
from itertools import combinations
from collections import Counter

count = Counter()

for row in df11['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list,2)))
    
print(count)
```

```python
count.most_common(10)
```

```python
for key,value in count.most_common(10):
    print(key,value)
```
